```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
names_by_year = {}
for year in range(1900, 2000, 1):
    names_by_year[year] = pd.read_csv(
        '/Users/nicolas13/Lessons_Netology/Lesson4_1/names/yob{}.txt'.format(year),
        names=['Name','Gender','Count']
    )
names_all = pd.concat(names_by_year, names=['Year', 'Pos'])
```

```python
gender_dynamics = names_all.groupby([names_all.index.get_level_values(0),'Gender']).sum().unstack('Gender')
```

```python
gender_dynamics.plot()
```
